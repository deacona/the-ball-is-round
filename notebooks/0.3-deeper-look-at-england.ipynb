# Deeper look at England 2006/7

Carrying out some more reconciliation using the SkySports Football Yearbook 2007-2008

```python
import sys
import pandas as pd
pd.set_option('display.max_columns', 500)
pd.set_option('display.expand_frame_repr', False)
import numpy as np
import warnings
warnings.filterwarnings('ignore')
sys.path.insert(0, r'..\src')
import clubs_wrangle
import utilities
```

```python
fulldata = utilities.get_master("fulldata")
```

## Middlesbrough

```python
base_filters = {
        "Season": ["2006-2007"],
        "Div": ["E0"],
        "Team": ["Middlesbrough"]
        }
boro = fulldata.copy()
for field, vals in base_filters.items():
    boro = boro[(boro[field].isin(vals))]

boro["Final score"] = boro["Goals"].astype(int).astype(str)+"-"+boro["GoalsOpp"].astype(int).astype(str)
boro["H/T score"] = boro["Goals1stHalf"].astype(int).astype(str)+"-"+boro["Goals1stHalfOpp"].astype(int).astype(str)

boro["DateDate"] = pd.to_datetime(boro["Date"])
import calendar
boro["MonthInt"] = boro.DateDate.dt.month
boro["Month"] = boro["MonthInt"].apply(lambda x: calendar.month_abbr[x])
boro["Day"] = boro["DateDate"].dt.day
    
# boro.info()
boro[["Date", "GameWeek", "Month", "Day", "HomeAway", "TeamOpp", "Result",
      "Final score", "H/T score"]].sort_values("Date")
```

Spot on!
